```python
# Test Accuracy

test=[0.9939393939393939,0.9924,0.9075757575757576,0.9833333333333333]
label=["Random Forest","Naive Bayes","Decision Tree","SVM"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=[6,4],dpi = 100)
plt.title('Test Accuracies')
plt.xlabel('Accuracy')
plt.ylabel('Type')
sns.barplot(x = test, y = label,palette="dark")
# plt.xlim(0.5,1)
```

```python
# CV Accuracy
cv=[0.9927272727272728,0.9950,0.9154545454545454,0.9827272727272728]
label=["Random Forest","Naive Bayes","Decision Tree","SVM"]
```

```python
plt.figure(figsize=[6,4],dpi = 100)
plt.title('Cross-Validation Accuracies')
plt.xlabel('Accuracy')
plt.ylabel('Type')
sns.barplot(x = cv,y = label,palette="dark")
# plt.xlim(0.5,1)
```

* Random Forest achieved the highest test accuracy (0.9939), very close to its cross-validation performance.

* Naive Bayes had a slightly better cross-validation score (0.9950), but a marginally lower test accuracy compared to Random Forest.

* SVM also performed well, with high scores in both phases.

* Decision Tree showed the lowest performance among the four models, indicating potential overfitting or lower generalization capability.

Random Forest was selected as the final model due to its excellent balance between cross-validation and test accuracy, showcasing both robustness and generalization.
